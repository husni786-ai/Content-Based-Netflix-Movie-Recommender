```python
df.set_index('show_id', inplace=True)
```

```python
df = df[df['type'] == 'Movie']  # optional, if you want only movies
```

```python
df['title'] = df['title'].str.lower()
```

```python
df['director'] = df['director'].str.lower()
df['cast'] = df['cast'].str.lower()
df['country'] = df['country'].str.lower()
```

```python
df['director'] = df['director'].str.strip()
df['cast'] = df['cast'].str.strip()
df['country'] = df['country'].str.strip()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
```

```python
df['rating'] = df['rating'].str.lower()
```

```python
df['duration'] = df['duration'].str.lower().str.replace(' min', '')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
```

```python
df['listed_in'] = df['listed_in'].str.lower()
```

```python
df['listed_in'] = df['listed_in'].str.replace(', ', ' ')
```

```python
df['description'] = df['description'].str.lower()
df['description'] = df['description'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
df['tags'] = df['title'] + ' ' + df['director'] + ' ' + df['cast'] + ' ' + df['listed_in'] + ' ' + df['description']
```

```python
df['tags'] = df['tags'].str.lower()              # Lowercase
df['tags'] = df['tags'].str.replace('[^a-zA-Z]', ' ', regex=True)  # Remove punctuation/numbers
df['tags'] = df['tags'].str.replace('\s+', ' ', regex=True)        # Remove extra spaces
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
# First, filter only movies (once at the start of your pipeline)
df_movies = df[df['type'] == 'Movie'].reset_index(drop=True)

# Clean titles
df_movies['title'] = df_movies['title'].str.lower()

# Choose a movie name
movie_name = 'Inception'.lower()

# Check if movie exists
if movie_name in df_movies['title'].values:
    movie_index = df_movies[df_movies['title'] == movie_name].index[0]

    # Now similarity[movie_index] will work properly
    distances = list(enumerate(similarity[movie_index]))
    sorted_movies = sorted(distances, key=lambda x: x[1], reverse=True)

    print("Top 5 similar movies to", movie_name.title(), "are:\n")
    for i in sorted_movies[1:6]:
        print(df_movies.iloc[i[0]]['title'].title())
else:
    print("Movie not found in dataset.")
```
